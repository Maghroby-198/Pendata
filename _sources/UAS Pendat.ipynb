# Preprosessing Data

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# Membuat DataFrame dari data
df = pd.DataFrame(data)
```

```python
# hapus kolom klasifikasi
X = df.drop(columns=["Provinsi","Tahun"])
```

```python
X
```

```python
# Melakukan Min-Max Scaling pada data suhu dan kelembapan
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(X)
```

```python
# Membuat DataFrame baru dari data yang telah di-skala
scaled_X = pd.DataFrame(scaled_data, columns=['Produksi (Scaled)',	'Luas Panen (Scaled)',	'Curah hujan (Scaled)',	'Kelembapan (Scaled)',	'Suhu rata-rata (Scaled)'])
```

```python
scaled_X
```

```python
from sklearn.decomposition import PCA
```

```python
# Melakukan reduksi dimensi dengan PCA
pca = PCA(n_components=1)  # Menentukan jumlah komponen yang diinginkan (misalnya 1)
reduced_data = pca.fit_transform(scaled_X)
```

```python
#  Membuat DataFrame baru dari data yang telah direduksi dimensinya
reduced_df = pd.DataFrame(reduced_data, columns=['Dimensi Baru'])
```

```python
# Menampilkan data yang telah direduksi dimensinya
reduced_df
```
