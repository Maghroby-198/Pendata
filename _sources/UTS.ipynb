# **Naive Bayes**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# input data
Cryotherapy = pd.read_csv("https://raw.githubusercontent.com/datasets/breast-cancer/master/data/breast-cancer.csv")
# Menampilkan data
Cryotherapy.head()
```

```python
# menampilkan informasi data
Cryotherapy.info()
```

```python
# Mengecek apakah ada deret yang kosong
Cryotherapy.empty
```

```python
# Melihat ukuran dari data
Cryotherapy.size
```

```python
# Variabel independen
x = Cryotherapy.drop(['deg-malig'], axis = 1)
x.head()
```

```python
# Variabel dependen
y = Cryotherapy['deg-malig']
y.head()
```

```python
# Import train_test_split function
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0, stratify=y)
```

```python
# Import Gaussian Naive Bayes model
from sklearn.naive_bayes import GaussianNB
# Mengaktifkan/memanggil/membuat fungsi klasifikasi Naive bayes
modelnb = GaussianNB()
x_train = modelnb.fit_transform(x_train)
x_test = modelnb.transform(x_test)
x_train[:5]
```
